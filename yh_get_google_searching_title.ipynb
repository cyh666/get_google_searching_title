```python
from selenium import webdriver
driver = webdriver.Chrome()
driver.get('http://www.google.com/')
q=driver.find_element_by_name('q')
q.send_keys('mask reticle')  # Search web
from selenium.webdriver.common.keys import Keys
q.send_keys(Keys.RETURN)      #Enter

# driver.find_element_by_css_selector('#rso a:nth-child(1)').click()

from bs4 import BeautifulSoup

def get_title():
    soup=BeautifulSoup(driver.page_source,'lxml')
    for ele in soup.select('#rso h3'):
            print(ele.text)

get_title()

for i in range(3,15):       # 頁數範圍
    q=driver.find_element_by_xpath(f'//*[@id="xjs"]/table/tbody/tr/td[{i}]/a')
    q.send_keys(Keys.RETURN)
    get_title()
            
```
